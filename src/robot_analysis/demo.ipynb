```python
from planar_nr import (
    PlanarRobotNR,
    animate_all_dynamics,
    plot_final_kinematics,
)
from sympy.utilities.lambdify import lambdify
import numpy as np
from IPython.display import HTML
import matplotlib as mpl

# %matplotlib notebook
mpl.rcParams['animation.embed_limit'] = 100  # đơn vị: MB
```

```python
n_links = 1
links = {
    'l': [1] * n_links,
    'm': [1.0] * n_links,
    'b': [1] * n_links,
}
robot = PlanarRobotNR(links, n_max=200)
```

```python
data = robot.simulate(
    x0=np.array([0.0, 0.0]),
    method='velocity',
    theta_sp=np.array([np.pi / 6]),
    vel_d=0,
    kp=0.001,
    ki=0.001,
    kd=0.0,
    tau_max=100.0,
)

plot_final_kinematics(**data)
```

```python
data = robot.simulate(
    x0=np.array([0.0, 0.0]),
    method='openloop',
    t1=0.2,
    t2=0.8,
    tf=1,
    theta_d=np.array([np.pi / 6]),
    tau_max=13,
)

plot_final_kinematics(**data)
# animate_all_dynamics(**data)
```
